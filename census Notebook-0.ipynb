```sql
-- querying population of india 

SELECT SUM(population) as population from project..data
```

```sql
--quering avg growth % of india

SELECT AVG(growth)*100 as average_growth from [project 1]..data
```

```sql
--- querying avg growth % of states

SELECT state, AVG(growth)*100 as average_growth from [project 1]..data GROUP BY state
```

```sql
---- querying sex ration state wise and arranging it to lowest to highest

SELECT state,ROUND (AVG(Sex_Ratio),0) as average_sex_ratio from [project 1]..data GROUP BY state ORDER BY average_sex_ratio desc;
```

```sql
---querying states having avg literacy rate > 90

SELECT state,ROUND (AVG(Literacy),0) as average_literacy_ratio from [project 1]..data GROUP BY state HAVING ROUND (AVG(Literacy),0) >90;
```

```sql
---- querying states starting with letter a but ending with d

SELECT DISTINCT state FROM [project 1]..data WHERE State like 'A%' or State like '%D'
```

```sql
---joining two tables and counting the no. of males and females. in each state


select d.state,sum(d.males) total_males,sum(d.females) total_females from
(select c.district,c.state state,round(c.population/(c.sex_ratio+1),0) males, round((c.population*c.sex_ratio)/(c.sex_ratio+1),0) females from
(select a.district,a.state,a.sex_ratio/1000 sex_ratio,b.population from [project 1]..data a inner join project..data b on a.district=b.district ) c) d
group by d.state;

```

```sql

----- querying top 3 district with highest literacy for each state using window function  


select a.* from
(select district,state,literacy,rank() over(partition by state order by literacy desc) rnk from [project 1]..data) a

where a.rnk in (1,2,3)
```
